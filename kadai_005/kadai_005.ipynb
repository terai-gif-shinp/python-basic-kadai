変数を使って四則演算しよう

上辺：10cm  
下辺：20cm  
高さ：5cm  

公式：台形の面積 =（上辺＋下辺）× 高さ ÷ 2


下記の台形の面積の計算結果を出力するプログラムを記述してください。  
出力結果は「◯◯cm²」となるように単位をつけてください。  
台形の面積の計算結果は、変数に代入してプログラム内で使用してください。  

```python
top_side = 10
bottom_side= 20
height = 5

area = (top_side + bottom_side) * height /2

print(f"{area}cm²")
```
